```python
#partie 1
#je choisis le critere de division Gini Impurity car il me semble plus simple que l'entropie

def gini(y):
    if len(y) == 0:
        return 0 #pour pas avoir de probleme de division par 0

    classes, counts = np.unique(y, return_counts=True)
    impurity = 1
    for count in counts:
        p = count / len(y)
        impurity -= p ** 2
    return impurity
```
